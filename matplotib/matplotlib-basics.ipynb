```python
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Graficos são formados por eixos
# Gráficos são editáveis por conta de serem eixos
# figures são onde os gráficos são plotados
```

```python
#first plot
import matplotlib.pyplot as plt

#simple plot
plt.plot(x, y)
```

```python
#creating labels x, y and title
plt.xlabel('eixo x')
plt.xlabel('eixo y')
plt.title('Título do gráfico')
plt.plot(y, x)
```

```python
# função subplot -> Forma de criarmos 2 eixos dentro de uma figure
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(x,y, 'g*-')
```

```python
# Orientação à objetos com matplot
# Instanciação de uma figure
fig = plt.figure()

# Adicionando eixos à uma figure
# montado a partir de 4 valores
eixos = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Ordem => Esquerda, inferior, largura, altura
eixos.plot(x, y, 'blue')

# Estilizar o gráfico
eixos.set_xlabel('Label x')
eixos.set_ylabel('Label y')
eixos.set_title('Example title')
```

```python
# Utilizando 2 eixos aplicados à uma figure
# Instanciação de uma figure
fig = plt.figure()

# Adicionando eixos à uma figure
# montado a partir de 4 valores
eixo1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Ordem => Esquerda, inferior, largura, altura
eixo2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Ordem => Esquerda, inferior, largura, altura


# Estilizar o gráfico 1
eixo1.plot(x, y, 'blue')
eixo1.set_xlabel('Label x1')
eixo1.set_ylabel('Label y1')
eixo1.set_title('Example title1')

# Estilizar o gráfico 2
eixo2.plot(x, y, 'red')
eixo2.set_xlabel('Label x2')
eixo2.set_ylabel('Label y2')
eixo2.set_title('Example title2')
```

```python
# Utilizando a função subplots
# A diferença entre a função subplot são as maiores possibilidades de estilização

#desempacotamento de tuplas
tupla = (1,2)
a, b = tupla
print(f'valor de a: {a} | Valor de b: {b}')

#desempacotamentos podem ser utilizados com figures pois uma figure é formada por uma tupla contendo a figure e o eixo
fig, ax = plt.subplots()
```

```python
# Aprofundando a instanciação de plots
fig, ax = plt.subplots(nrows=1, ncols=2)

# Quando instanciamos mais de um eixo no plot, teremos uma lista contendo os eixos
ax

# Podemos de iterar sobre a lista pra poder manipular cada eixo da figure
for axis in ax:
    axis.plot(x,y)
    axis.set_title('Printando utlizando o for')
    
# Podemos também manipular os eixos acessando diretamento o índice do eixo
# Utilizando a função tight_layout podemos ajustar o tamanho, ou o figsize como parâmetro na criação do subplots
fig2, ax2 = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
ax2[0].plot(x,y, 'blue')
ax2[0].set_title('printando acessando o índice 0')
ax2[1].plot(x,y, 'red')
ax2[1].set_title('printando acessando o índice 1')

```

```python
#Quando trabalhamos com figures mais complexar, temos de retorno uma lista de listas
fig3, ax3 = plt.subplots(nrows=2, ncols=2)
fig3
ax3
```

```python
# Customização de gráficos com subplots
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='x**2') #Podemos criar legenda para o gráfico passando o parâmetro label
ax.plot(x, x**3, label='x**2')

ax.set_title('Customizando')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')

# Para printarmos uma legenda utilizamos o método legend
ax.legend() #Podermos definir onde a legenda ficará com o parâmetro loc

# lista com outros parâmetros possíveis::
# color
# alpha => Transparência
# linewidth ou lw
# linestyle ou ls
# marker 
# markersize
# marerfacecolor
# outros...
```

```python
# Customizando intevalor de dados dentro de um eixo
fig, axes = plt.subplots(1,3, figsize=(12,4))

axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title('Intervalo padrão')

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis('tight') # Definindo tamanho de gráfico comprimido
axes[1].set_title('Eixos apertados')

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0,60]) # Definindo tamanho de gráfico manualmente
axes[2].set_xlim([0,5]) # Definindo tamanho de gráfico manualmente
axes[2].set_title('Eixos definidos manualmente')
```
