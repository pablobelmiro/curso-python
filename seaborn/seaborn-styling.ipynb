Estilização de gráficos

```python
import seaborn as sns

#comando utilizando os datasets da lib
tips = sns.load_dataset('tips')
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips, palette='deep', hue='sex')
```

```python
# De maneira geral, para estilizações mais complexas, é utilizado a figure
# Exemplo:

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,8))
sns.countplot(data=tips, x='sex', palette='deep', ax=ax)

ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
```
