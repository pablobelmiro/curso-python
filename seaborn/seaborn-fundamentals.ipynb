```python
import seaborn as sns
```

TIPOS DE PLOTS

```python
#comando utilizando os datasets da lib
tips = sns.load_dataset('tips')

sns.distplot(tips['total_bill'])
```

```python
# Explorando dados::
# KDE => Estimativa de distribuição

#printando sem o KDE
sns.distplot(tips['total_bill'], kde=False)
```

```python
# Trabalhando com subdivisões
# O parametro bins define a subdivisão

sns.distplot(tips['total_bill'], kde=False, bins=200)
```

```python
# UTILIZANDO O MÉTODO joinplot()
# Ele analisa 2 variáveis simultâneamente
# Criamos 2 distplots + um terceiro comparando os dois gráficos

sns.jointplot(x='total_bill', y='tip', data=tips)
```

TRABALHANDO PARÂMETROS COM O joinplot()

```python
# controle de tamanhos
# kind = tipo de gráfico que será impresso
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

UTILIZANDO O MÉTODO pairplot()

```python
sns.pairplot(tips)

# podemos utiliza segregar columas definindo categorias, passando por parâmetro
sns.pairplot(tips, hue='tip')
```
