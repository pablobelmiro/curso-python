```python
import seaborn as sns

#comando utilizando os datasets da lib
tips = sns.load_dataset('tips')
```

```python
#plot puro
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
#utilizando markers
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['x', 'v'])
```

```python
# Quebrando gr√°fico em colunas de acordo com tempo
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='time', row='day',markers=['o', 'x'])
```
