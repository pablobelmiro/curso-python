PLOTS MATRICIAIS

```python
import seaborn as sns

#comando utilizando os datasets da lib
flights = sns.load_dataset('flights')
flights
```

Obs.: De maneira geral devemos organizar os dados para podermos realizarmos plots do tipo matricial, caso necessário, utilizar lib pandas

```python
dfFlights = flights.pivot_table(index='year', columns='month', values='passengers')
dfFlights
```

```python
sns.heatmap(dfFlights)
```

```python
#utilizando alguns parâmetros para poder plotar matrizes
#criando figure para ajustar a escala do gráfico + ajuste de eixo → Utilizar a lib matplotlib para isso

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,8))
sns.heatmap(dfFlights, ax=ax)
```

```python
# plotando valores da matriz distribuídos e formatados
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,8))
sns.heatmap(dfFlights, ax=ax, annot=True, fmt=".0f")
```

```python
# plotando valores da matriz distribuídos e formatados
sns.clustermap(data=dfFlights)
```
